```python
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.arange(-6, 6, .1)
y = torch.sin(x)
plt.plot(x, y, 'g--')
```

```python
trainx = x[(x >= -3) & (x <= 3)]
trainy = torch.sin(trainx) + torch.normal(0, 0.1, trainx.shape)
```

```python
# @title
class MLP(torch.nn.Module):
  def __init__(self):
    super().__init__()

    self.subnet1 = torch.nn.Linear(1, 16)
    self.subnet2 = torch.nn.Linear(16, 16)
    self.subnet3 = torch.nn.Linear(16, 1)

  def forward(self, x):
    x = torch.nn.functional.sigmoid(self.subnet1(x))
    x = torch.nn.functional.sigmoid(self.subnet2(x))
    x = self.subnet3(x)

    return x

```

```python
class SubNet(torch.nn.Module):
    def __init__(self, depth, width=16):
        super().__init__()
        # Creating a list of layers based on depth
        layers = []
        in_dim = 1
        for _ in range(depth):
            layers.append(torch.nn.Linear(in_dim, width))
            layers.append(torch.nn.Sigmoid())
            in_dim = width
        self.features = torch.nn.Sequential(*layers)
        self.output = torch.nn.Linear(width, 1)

    def forward(self, x):
        return self.output(self.features(x))

class ProposedNet(torch.nn.Module):
    def __init__(self):
        super().__init__()
        # Three subnets of varying length (1, 2, and 3 layers)
        self.path1 = SubNet(depth=1)
        self.path2 = SubNet(depth=2)
        self.path3 = SubNet(depth=3)

        # Learnable weights for the combination
        self.alpha = torch.nn.Parameter(torch.ones(3) / 3)

    def forward(self, x):
        # Linear combination as described in the paper
        return (self.alpha[0] * self.path1(x) +
                self.alpha[1] * self.path2(x) +
                self.alpha[2] * self.path3(x))
```

```python
standardMLP, proposal = MLP(), ProposedNet()
optimizer1 = torch.optim.Adam(standardMLP.parameters(), lr = 1e-3)
optimizer2 = torch.optim.Adam(proposal.parameters(), lr = 1e-3)
```

```python
standardMLP.train()
proposal.train()
loss_fn = torch.nn.MSELoss()
loss_memory = float("inf")
for e in range(10000+1):
  optimizer1.zero_grad()
  optimizer2.zero_grad()

  MLPLoss = loss_fn(standardMLP(trainx.unsqueeze(1)).squeeze(), trainy)
  proposalLoss = loss_fn(proposal(trainx.unsqueeze(1)).squeeze(), trainy)

  # Early Stop
  if e%250 == 0:
    if loss_memory < (MLPLoss.item()):
      print ("Loss :" + str(MLPLoss.item()))
      print ("Early Stop at epoch: " + str(e))
      break
    loss_memory = MLPLoss.item()

  MLPLoss.backward()
  proposalLoss.backward()
  optimizer1.step()
  optimizer2.step()

  if e%500 == 0:
    print ("Epoch :" + str(e) + " MLPLoss: " + str(MLPLoss.item()), " Proposal Loss" + str(proposalLoss.item()))
    with torch.no_grad():
      fig, ax = plt.subplots(1, 2, figsize = (10, 10))
      ax[0].plot(x, standardMLP(x.unsqueeze(1)), "--", label = 'Prediction')
      ax[0].plot(x, y)
      ax[0].scatter(trainx, trainy, s = 5, color = 'green', label = 'Training Data')
      ax[0].legend()
      ax[0].grid(True)
      ax[1].plot(x, proposal(x.unsqueeze(1)), "--", label = 'Prediction')
      ax[1].plot(x, y)
      ax[1].scatter(trainx, trainy, s = 5, color = 'green', label = 'Training Data')
      ax[1].legend()
      ax[1].grid(True)
      plt.show()
```
